```python
save_root = '/camp/lab/znamenskiyp/home/shared/presentations/Cosyne2023/ver'+str(VERSION)+'/depth_decoder/'
accs_same_closeloop_old2 = np.load(f'{save_root}accs_same_closeloop_old2.npy')
accs_bins_closeloop_old2 = np.load(f'{save_root}accs_bins_closeloop_old2.npy')
conmats_same_closeloop_old2 = np.load(f'{save_root}conmats_same_closeloop_old2.npy')
conmats_bins_closeloop_old2 = np.load(f'{save_root}conmats_bins_closeloop_old2.npy')
accs_same_openloop_old2 = np.load(f'{save_root}accs_same_openloop_old2.npy')
accs_bins_openloop_old2 = np.load(f'{save_root}accs_bins_openloop_old2.npy')
conmats_same_openloop_old2 = np.load(f'{save_root}conmats_same_openloop_old2.npy')
conmats_bins_openloop_old2 = np.load(f'{save_root}conmats_bins_openloop_old2.npy')

accs_same_closeloop_new3 = np.load(f'{save_root}accs_same_closeloop_new3.npy')
accs_bins_closeloop_new3 = np.load(f'{save_root}accs_bins_closeloop_new3.npy')
conmats_same_closeloop_new3 = np.load(f'{save_root}conmats_same_closeloop_new3.npy')
conmats_bins_closeloop_new3 = np.load(f'{save_root}conmats_bins_closeloop_new3.npy')
accs_same_openloop_new3 = np.load(f'{save_root}accs_same_openloop_new3.npy')
accs_bins_openloop_new3 = np.load(f'{save_root}accs_bins_openloop_new3.npy')
conmats_same_openloop_new3 = np.load(f'{save_root}conmats_same_openloop_new3.npy')
conmats_bins_openloop_new3 = np.load(f'{save_root}conmats_bins_openloop_new3.npy')


def bar_plot_ttest(group1, group2, fig, ax, labels):
    # Calculate means and standard deviations
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1), np.std(group2)

    # Calculate t-test
    from scipy.stats import ttest_ind
    t_statistic, p_value = ttest_ind(group1, group2)

    # Plot the data
    bar1 = ax.bar(0, mean1, yerr=std1, capsize=10, width = 0.25, label=labels[0])
    bar2 = ax.bar(0.3, mean2, yerr=std2, capsize=10, width = 0.25, label=labels[1])

    # Add bracket and asterisk if significant
    if p_value < 0.05:
        star = '*'
        if p_value < 0.01:
           star = '**' 
        elif p_value < 0.001:
           star = '***' 
        # Add square bracket
        x0, x1 = bar1[0].get_x() + bar1[0].get_width() / 2, bar2[0].get_x() + bar2[0].get_width() / 2
        y, h, col = max([mean1 + std1]), 0.05, 'k'
        ax.plot([x0, x0, x1, x1], [y+h, y+2*h, y+2*h, y+h], lw=1.5, c=col)
        # Add asterisk on top of bracket
        ax.text((x0+x1)*.5, y+2*h,star, ha='center', va='bottom', fontsize=24)

    # Set the x-axis labels and title
    ax.set_xticks([0, 0.3])
    ax.set_xticklabels([labels[0], labels[1]])
    # ax.set_title("Comparison of Two Groups")

    despine()

fontsize_dict = {
    'title': 10,
    'xlabel': 12,
    'ylabel': 12,
    'xticks': 10,
    'yticks': 10,
    'legend': 10, 
    'text': 10
}   
fig = plt.figure(figsize=(6,5))
ax = plt.subplot(1,2,1)
bar_plot_ttest(group1=np.array(accs_same_closeloop_old2).flatten(), 
               group2=np.array(accs_same_openloop_old2).flatten(), fig=fig, ax=ax, labels=['Closed loop', 'Open loop'])
ax.axhline(y=1/5, color='r', linestyle='dotted', linewidth=2)
plt.ylabel('Classifier accuracy', fontsize=fontsize_dict['xlabel'])
plt.xticks(fontsize=fontsize_dict['xticks'])
plt.yticks(fontsize=fontsize_dict['yticks'])
# plt.legend(loc='upper right',bbox_to_anchor=(1, 0.8), fontsize=fontsize_dict['legend'])
ylim = ax.get_ylim()

ax = plt.subplot(1,2,2)
bar_plot_ttest(group1=np.array(accs_same_closeloop_new3).flatten(), 
               group2=np.array(accs_same_openloop_new3).flatten(), fig=fig, ax=ax, labels=['Closed loop', 'Open loop'])
ax.axhline(y=1/8, color='r', linestyle='dotted', linewidth=2)
plt.ylim(ylim)
plt.ylabel('Classifier accuracy', fontsize=fontsize_dict['xlabel'])
plt.xticks(fontsize=fontsize_dict['xticks'])
plt.yticks(fontsize=fontsize_dict['yticks'])
# plt.legend(loc='upper right',bbox_to_anchor=(1, 0.8), fontsize=fontsize_dict['legend'])
plt.tight_layout(pad=1)
plt.savefig(save_root+'accuracy_all.pdf')
```

```python
def plot_confusion_matrix(conmat, fontsize_dict, vmin, vmax, k=1):
    plt.imshow(conmat, interpolation='nearest', cmap='Blues', vmin=vmin, vmax=vmax)
    fmt = 'd'
    thresh = conmat.max() *4/5.
    for i, j in itertools.product(range(conmat.shape[0]), range(conmat.shape[1])):
        plt.text(j, i, str(int(conmat[i, j])*k),
                horizontalalignment="center",
                color="white" if conmat[i, j] > thresh else 
                "black", fontsize=fontsize_dict['text'])
    plt.xlabel('Predicted depth class',fontsize=fontsize_dict['xlabel'])
    plt.ylabel('True depth class',fontsize=fontsize_dict['ylabel'])

fontsize_dict = {
    'title': 10,
    'xlabel': 12,
    'ylabel': 12,
    'xticks': 10,
    'yticks': 10,
    'legend': 10, 
    'text': 10
}   
plt.figure(figsize=(10,5))
conmat_same_closeloop_ave_old2 = np.average(conmats_same_closeloop_old2, axis=0).squeeze()
conmat_same_openloop_ave_old2 = np.average(conmats_same_openloop_old2, axis=0).squeeze()
vmax = np.nanmax((conmat_same_closeloop_ave_old2,conmat_same_openloop_ave_old2))
vmin = np.nanmin((conmat_same_closeloop_ave_old2,conmat_same_openloop_ave_old2))
plt.subplot(121)
plot_confusion_matrix(conmat_same_closeloop_ave_old2, fontsize_dict, k=conmats_same_closeloop_old2.shape[0], vmin=vmin, vmax=vmax)
plt.title('Closed loop', fontsize=fontsize_dict['title'])


plt.subplot(122)
plot_confusion_matrix(conmat_same_openloop_ave_old2, fontsize_dict, k=conmats_same_closeloop_old2.shape[0], vmin=None, vmax=None)
plt.title('Open loop', fontsize=fontsize_dict['title'])
plt.savefig(save_root+'conmat_all_old2.pdf')

plt.figure(figsize=(10,5))
conmat_same_closeloop_ave_new3 = np.average(conmats_same_closeloop_new3, axis=0).squeeze()
conmat_same_openloop_ave_new3 = np.average(conmats_same_openloop_new3, axis=0).squeeze()
vmax = np.nanmax((conmat_same_closeloop_ave_new3,conmat_same_openloop_ave_new3))
vmin = np.nanmin((conmat_same_closeloop_ave_new3,conmat_same_openloop_ave_new3))
plt.subplot(121)
plot_confusion_matrix(conmat_same_closeloop_ave_new3, fontsize_dict, k=conmats_same_closeloop_new3.shape[0], vmin=vmin, vmax=vmax)
plt.title('Closed loop', fontsize=fontsize_dict['title'])


plt.subplot(122)
plot_confusion_matrix(conmat_same_openloop_ave_new3, fontsize_dict, k=conmats_same_closeloop_new3.shape[0], vmin=None, vmax=None)
plt.title('Open loop', fontsize=fontsize_dict['title'])
plt.savefig(save_root+'conmat_all_new3.pdf')
```
